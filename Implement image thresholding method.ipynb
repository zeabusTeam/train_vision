### Your method

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

image = cv2.imread('./images/01.jpeg')
img = cv2.cvtColor(image,cv2.COLOR_RGB2GRAY)
w,l = img.shape
```

```python
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)
```

```python
_, th = cv2.threshold(img,200,255,cv2.THRESH_BINARY)
```

```python
plt.imshow(th,cmap='gray')
```

```python
def thresholding(array,a,b):
    w,l = array.shape
    result = np.zeros([w,l])
    for x in range(w):
        for y in range(l):
            p = int(array[x:(x+1),y:(y+1)])
            if (a < p < b):
                result[x,y] = 255
            else:
                result[x,y] = 0
    return np.array(result,dtype='uint8')
```

```python
img_out = thresholding(img,200,255)
```

```python
plt.imshow(img_out,cmap='gray')
```
